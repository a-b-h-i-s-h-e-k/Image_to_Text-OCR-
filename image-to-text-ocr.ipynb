```python
# Importing the pytesseract library for Optical Character Recognition (OCR)
import pytesseract

# Convert the image to 'RGB' mode before processing
# pytesseract works best when the image is in RGB mode. This ensures that the image is in a format that Tesseract can easily process.
text_data = pytesseract.image_to_string(img.convert('RGB'), lang='eng')  
# 'img.convert('RGB')' converts the image to RGB mode. This is often required because certain image formats (like grayscale) may not be handled properly by OCR tools.
# 'lang="eng"' specifies that the OCR should use the English language model for text recognition.

# Print the extracted text
# The result from 'image_to_string' will be the text recognized from the image, which can be printed or used for further processing.
print(text_data)


# image_to_string() is called on the image object (img), and the function will return any readable text found in the image.
# This can then be processed, stored, or used in further applications.
```

## Extract Specific Fields

```python
# Importing the regular expression (re) module to work with patterns in strings
import re

# Use re.search to find the pattern "Name: (\w+)" in the extracted text data
m = re.search("Name: (\w+)", text_data)
# re.search: This function searches for the first occurrence of the specified pattern in the text.
# - "Name: (\w+)": This is the regular expression pattern we are searching for.
#     - "Name: ": Looks for the literal string "Name: ".
#     - (\w+): A capturing group that matches one or more word characters (letters, numbers, or underscores).
#       This will capture the part of the text that immediately follows "Name: ".

# If the pattern is found, 'm' will contain a match object. We extract the first group (the name)
name = m[1]  # m[1]: Retrieves the text captured by the first capturing group (\w+), which corresponds to the name.

# Output the extracted name
name  # This will display the extracted name from the text if the pattern was found.
```

```python
# Importing the regular expression module in Python
import re

# Using the search function from the re module to search for a specific pattern
# The pattern "Start Date: (\S+)" looks for "Start Date: " followed by one or more non-whitespace characters
m = re.search("Start Date: (\S+)", text_data)

# If a match is found, 'm' will contain the match object.
# The [1] refers to the first capturing group (the part inside the parentheses), which will be the non-whitespace characters after "Start Date: ".
# This line assigns the matched value to the variable 'start_date'.
start_date = m[1]

# Printing the matched start date
start_date
```

```python
# Importing the regular expression module in Python
import re

# Using the search function from the re module to search for a specific pattern
# The pattern "Geo-Coordinates: (\S+)" looks for "Geo-Coordinates: " followed by one or more non-whitespace characters
m = re.search("Geo-Coordinates: (\S+)", text_data)

# If a match is found, 'm' will contain the match object.
# The [1] refers to the first capturing group (the part inside the parentheses), which will be the non-whitespace characters after "Geo-Coordinates: ".
# This line assigns the matched value (coordinates) to the variable 'coordinates'.
coordinates = m[1]

# Printing the matched geo-coordinates
coordinates
```
